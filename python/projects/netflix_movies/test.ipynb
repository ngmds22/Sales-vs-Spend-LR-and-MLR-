```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('netflix_titles.csv', encoding='latin1')

# Display the first few rows
print(df.head())

# Get information about the dataset
print(df.info())

# Get summary statistics of the numerical columns
print(df.describe())
```

```python
import matplotlib.pyplot as plt

# Count the number of movies and TV shows
show_counts = df['type'].value_counts()

# Create a bar chart
plt.bar(show_counts.index, show_counts.values)

# Add labels and title
plt.xlabel('Show Type')
plt.ylabel('Count')
plt.title('Number of Movies and TV Shows')

# Display the chart
plt.show()

# Count the number of movies and TV shows released each year
release_counts = df['release_year'].value_counts().sort_index()

# Create a line chart
plt.plot(release_counts.index, release_counts.values)

# Add labels and title
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.title('Number of Movies and TV Shows Released Each Year')

# Display the chart
plt.show()
```
