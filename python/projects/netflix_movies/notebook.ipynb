```python
# import pandas
import pandas as pd
```

```python
df = pd.read_csv('netflix_titles.csv', encoding='latin1')
df.columns
```

```python
df = df[df.columns[:12]]
df.columns
```

```python
# Content-type distribution
# How is content distributed between movies and tv shows on netflix?

df.type.value_counts()
```

```python
# Release year analysis
# What trends can be observed in the release years of content added to Netflix? Has there been an increase
# in recent content over the years?
(
    df.release_year
      .value_counts()
      .sort_index()
      .plot(kind='line')
)
```

```python
df.dropna(subset=['date_added'], inplace=True)
df.date_added = df.date_added.str.strip()
df.date_added = pd.to_datetime(df.date_added)

(
    df.date_added
      .dt.year
      .value_counts()
      .sort_index()
      .plot(kind='line')
)
```
